```python
datos = pd.read_excel("LMC_VI_filtrada.xlsx") #ya están filtradas por OGLE4, LMC, Cefeidas de tipo fundamental únicamente
datos
```

```python
mag_I = datos["I0mean"].dropna()
len(mag_I)
#mag_V = datos["V0mean"].dropna()
#len(mag_V)
```

```python
resta = datos['I'].subtract(datos['A_I'])
resta.dropna()
```

```python
datos_I = datos[["I0mean", "P_0"]].dropna()
datos_I
```

```python
datos_V = datos[["V0mean", "P_0"]].dropna()
datos_V
```

```python
mag_I = datos_I["I0mean"]
log_pi = np.log10(datos_I["P_0"])
mag_V = datos_V["V0mean"]
log_pv = np.log10(datos_V["P_0"])
```

```python
from scipy.stats import linregress

# Perform Least Squares Regression
slope, intercept, r_value, p_value, slope_err = linregress(log_pi, mag_I)
intercept_err = slope_err * np.sqrt(np.sum(log_pi**2) / len(log_pi))  # Calculate error for intercept


# Print the results
print(f"Slope (m): {slope:.4f} ± {slope_err:.4f}")
print(f"Intercept (b): {intercept:.4f} ± {intercept_err:.4f}")
print(f"R-squared: {r_value**2:.4f}")

# Generate regression line
log_p_range = np.linspace(min(log_pi), max(log_pi), 100)
best_fit_line = slope * log_p_range + intercept

# Plot Data & Best-Fit Line
plt.figure(figsize=(8, 6))
plt.scatter(log_pi, mag_I, color='purple', alpha=0.6, edgecolor='black', label="Data")
plt.plot(log_p_range, best_fit_line, color='red', linewidth=2, label=f"Fit: I0 = ({slope:.2f} ± {slope_err:.2f}) Log P + ({intercept:.2f} ± {intercept_err:.2f})")

plt.gca().invert_yaxis()  # Invert y-axis for magnitudes
plt.xlabel('Log P')
plt.ylabel('Magnitud I')
plt.grid(True, linestyle="--", alpha=0.5)
#plt.title('Least Squares Regression: I vs Log P for LMC')
plt.legend()

# Show the plot
plt.show()
```

```python
from scipy.stats import linregress

# Perform Least Squares Regression
slope, intercept, r_value, p_value, slope_err = linregress(log_pv, mag_V)
intercept_err = slope_err * np.sqrt(np.sum(log_pv**2) / len(log_pv))  # Calculate error for intercept


# Print the results
print(f"Slope (m): {slope:.4f} ± {slope_err:.4f}")
print(f"Intercept (b): {intercept:.4f} ± {intercept_err:.4f}")
print(f"R-squared: {r_value**2:.4f}")

# Generate regression line
log_p_range = np.linspace(min(log_pv), max(log_pv), 100)
best_fit_line = slope * log_p_range + intercept

# Plot Data & Best-Fit Line
plt.figure(figsize=(8, 6))
plt.scatter(log_pv, mag_V, color='blue', alpha=0.6, edgecolor='black', label="Data")
plt.plot(log_p_range, best_fit_line, color='red', linewidth=2, label=f"Fit: V0 = ({slope:.2f} ± {slope_err:.2f}) Log P + ({intercept:.2f} ± {intercept_err:.2f})")

plt.gca().invert_yaxis()  # Invert y-axis for magnitudes
plt.xlabel('Log P')
plt.ylabel('Magnitud V')
plt.grid(True, linestyle="--", alpha=0.5)
#plt.title('Least Squares Regression: V vs Log P for LMC')
plt.legend()

# Show the plot
plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan, het_white
```

```python
# Perform regression analysis using statsmodels for heteroscedasticity tests
X = sm.add_constant(log_pv)  # Add a constant for the intercept
model = sm.OLS(mag_V, X).fit()
# Perform the Breusch-Pagan test
bp_test = het_breuschpagan(model.resid, model.model.exog)
bp_labels = ['LM Statistic', 'LM-Test p-value', 'F-Statistic', 'F-Test p-value']
bp_results = dict(zip(bp_labels, bp_test))
print("Breusch-Pagan Test Results:")
print(bp_results)
# Perform the White test
white_test = het_white(model.resid, model.model.exog)
white_labels = ['Lagrange Multiplier statistic', 'LM-test p-value', 'F-statistic', 'F-test p-value']
white_results = dict(zip(white_labels, white_test))
print("White Test Results:")
print(white_results)
```
