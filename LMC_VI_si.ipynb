```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("ogle_3_4_cep.csv",index_col=[0,1,3,2]).sort_index()
```

```python
datos
```

```python
ola = datos.loc["ogle4",'lmc','F'][['mean I','mean V','P_0']].dropna()
```

```python
ola
```

```python
mean_v = ola["mean V"]
mean_v
```

```python
mean_I = ola["mean I"]
mean_I
periodos = ola["P_0"]
log_p = np.log10(periodos)
```

```python
mean_I
len(mean_v)
```

```python
from scipy.stats import linregress

# Perform Least Squares Regression
slope, intercept, r_value, p_value, slope_err = linregress(log_p, mean_I)
intercept_err = slope_err * np.sqrt(np.sum(log_p**2) / len(log_p))  # Calculate error for intercept


# Print the results
print(f"Slope (m): {slope:.4f} ± {slope_err:.4f}")
print(f"Intercept (b): {intercept:.4f} ± {intercept_err:.4f}")
print(f"R-squared: {r_value**2:.4f}")

# Generate regression line
log_p_range = np.linspace(min(log_p), max(log_p), 100)
best_fit_line = slope * log_p_range + intercept

# Plot Data & Best-Fit Line
plt.figure(figsize=(8, 6))
plt.scatter(log_p, mean_I, color='purple', alpha=0.6, edgecolor='black', label="Data")
plt.plot(log_p_range, best_fit_line, color='red', linewidth=2, label=f"Fit: J = ({slope:.2f} ± {slope_err:.2f}) Log P + ({intercept:.2f} ± {intercept_err:.2f})")

plt.gca().invert_yaxis()  # Invert y-axis for magnitudes
plt.xlabel('Log P')
plt.ylabel('Magnitud I')
plt.grid(True, linestyle="--", alpha=0.5)
plt.title('Least Squares Regression: I vs Log P for LMC')
plt.legend()

# Show the plot
plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan, het_white
```

```python
# Perform regression analysis using statsmodels for heteroscedasticity tests
X = sm.add_constant(log_p)  # Add a constant for the intercept
model = sm.OLS(mean_v, X).fit()
# Perform the Breusch-Pagan test
bp_test = het_breuschpagan(model.resid, model.model.exog)
bp_labels = ['LM Statistic', 'LM-Test p-value', 'F-Statistic', 'F-Test p-value']
bp_results = dict(zip(bp_labels, bp_test))
print("Breusch-Pagan Test Results:")
print(bp_results)
# Perform the White test
white_test = het_white(model.resid, model.model.exog)
white_labels = ['Lagrange Multiplier statistic', 'LM-test p-value', 'F-statistic', 'F-test p-value']
white_results = dict(zip(white_labels, white_test))
print("White Test Results:")
print(white_results)
```

```python
from astropy.stats import sigma_clip
```

```python
filtered_data = sigma_clip(mean_v, sigma=2.5, maxiters=10)
```

```python
filtered_data
```

```python
plt.plot(log_p, mean_v, '+', color='#1f77b4', label="original data")
plt.plot(log_p[filtered_data.mask], mean_v[filtered_data.mask], 'x',
         color='#d62728', label="rejected data")
plt.gca().invert_yaxis()  # Invert y-axis for magnitudes
plt.xlabel('LogP')
plt.ylabel('V')
plt.grid()
plt.legend(loc=2, numpoints=1)
```

```python
#Mínimos cuadrados con datos filtrados
slope, intercept, r_value, p_value, slope_err = linregress(log_p[filtered_data.mask], mean_v[filtered_data.mask])
intercept_err = slope_err * np.sqrt(np.sum(log_p[filtered_data.mask]**2) / len(log_p[filtered_data.mask]))  # Calculate error for intercept


# Print the results
print(f"Slope (m): {slope:.4f} ± {slope_err:.4f}")
print(f"Intercept (b): {intercept:.4f} ± {intercept_err:.4f}")
print(f"R-squared: {r_value**2:.4f}")
```
